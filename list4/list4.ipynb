## Wprowadzenie
Skrypt pokazuje jak użyć pakietu SciKit do grupowania danych. Rozważane są dwa przykłady: dwuwymiarowe dane wygenerowane losowo z mieszaniny rozkładów gaussowskich oraz dane IRIS.

```python
% matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.metrics as metrics

from sklearn.cluster import KMeans, MiniBatchKMeans, Birch, DBSCAN
```

```python
import matplotlib.colors as colors
import math
from itertools import cycle
from scipy.spatial import distance

def dunn_index(X, labels, plt_K):
    WCD = []
    BCD = []
    if((-1) in labels):
        plt_K -= 1
    R = np.zeros((plt_K,2))
    for k in range(plt_K):
        mask = (labels == k)
        A = X[mask,:]
        C = A.shape[0]
        distA = distance.cdist(A,A,'euclidean')
        WCD.append(sum(sum(distA))/(C*(C-1)))
        R[k] = sum(A)/C
    distR = distance.cdist(R,R,'euclidean')
    if(plt_K > 1):
        distR += np.diag(np.ones(plt_K)*(1e500))
    return(np.min(distR)/max(WCD))
    
def davies_bouldin_index(X, labels, plt_K):
    if((-1) in labels):
        plt_K -= 1
    r = np.zeros((plt_K,2))
    S = np.zeros((plt_K))
    R = np.zeros((plt_K,plt_K))
    for k in range(plt_K):
        mask = (labels == k)
        A = X[mask,:]
        C = A.shape[0]
        r[k] = sum(A)/C
        S[k] = math.sqrt(sum(np.sum((A-r[k])**2,axis=1))/C)
    M = distance.cdist(r,r,'euclidean')
    M += np.diag(np.ones(plt_K)*(1e500))
    R += S+S[np.newaxis,:].T
    D = R/M
    return(sum(np.amax(D,axis=1))/plt_K)
    
def plotClustering(X, plt_labels):
    plt_colors = cycle(colors.cnames.keys())
    for i in range(30):
        next(plt_colors)
    plt_K = np.unique(plt_labels).size
    mask = (plt_labels == -1)
    plt.plot(X[mask,0], X[mask,1], 'k.')
    for k in range(plt_K):
        color = next(plt_colors)
        mask = (plt_labels == k)
        plt.plot(X[mask, 0], X[mask, 1], 'w', markerfacecolor=color, marker='o')
    
    plt.show()
    print("Silhouette Coefficient:",metrics.silhouette_score(X, plt_labels, metric='euclidean'))
    print("Dunn Index:",dunn_index(X,plt_labels,plt_K))
    print("Davies Bouldin Index:",davies_bouldin_index(X,plt_labels,plt_K))
```

## 1. Pierwszy zestaw danych
Dwuwymiarowe dane wygenerowane losowo z mieszaniny rozkładów gaussowskich

```python
from sklearn import datasets

centers_ = [[1, 1], [3, 3], [5, 1]]
X, labels = datasets.make_blobs(n_samples=3000, n_features=2, centers=centers_, cluster_std=0.5)

plotClustering(X, labels)
```

### Grupowanie za pomocą KMeans

```python
for i in np.arange(1,4):
    k_means = KMeans(init='k-means++', n_clusters=3, n_init=i)
    k_means.fit(X)

    plotClustering(X, k_means.labels_)
    
```

### Grupowanie za pomocą MiniBatchKMeans

```python
mini_batch_k_means = MiniBatchKMeans(init='k-means++', n_clusters=3, batch_size=100, n_init=1, max_no_improvement=10, verbose=0, random_state=0)
mini_batch_k_means.fit(X)

plotClustering(X, mini_batch_k_means.labels_)
```

### Grupowanie za pomocą Birch bez wtórnego grupowania

```python
for i in np.arange(0.2,1.,0.25):
    birch = Birch(threshold=i, n_clusters=None)
    birch.fit(X)

    plotClustering(X, birch.labels_)
```

### Grupowanie za pomocą Birch z wtórnym grupowaniem

```python
for i in np.arange(0.2,1.,0.25):
    birch = Birch(threshold=i, n_clusters=3)
    birch.fit(X)

    plotClustering(X, birch.labels_)
    
```

### Grupowanie za pomocą DBScan

```python
for i in np.arange(0.1,0.5,0.05):
    dbscan = DBSCAN(eps=i, min_samples=25)
    dbscan.fit(X)

    plotClustering(X, dbscan.labels_)
    
```

```python
def plotClusteringBirch(X, plt_labels, centers):
    plt_colors = cycle(colors.cnames.keys())
    for i in range(10):
        next(plt_colors)
    plt_K = np.unique(plt_labels).size
    for k in range(plt_K):
        color = next(plt_colors)
        mask = (plt_labels == k)
        plt.plot(X[mask, 0], X[mask, 1], 'w', markerfacecolor=color, marker='o')
    plt.plot(centers[:,0],centers[:,1], 'kD')
    plt.show()

centers_ = [[1, 1], [3, 3], [5, 1]]
X, labels = datasets.make_blobs(n_samples=3000, n_features=2, centers=centers_, cluster_std=0.5)

plotClustering(X, labels)
```

```python
birch = Birch(threshold=0.25, n_clusters=3)
birch.fit(X)
plotClusteringBirch(X,birch.labels_,birch.root_.centroids_)
```

```python
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data

def normalization(X):
    d = X.shape[1]
    A = np.zeros(X.shape)
    for i in range(d):
        min_ = np.min(X[:,i])
        max_ = np.max(X[:,i])
        A[:,i] = (X[:,i] - min_)/(max_-min_)
    return A

def standarization(X):
    d = X.shape[1]
    A = np.zeros(X.shape)
    for i in range(d):
        mean_ = np.mean(X[:,i])
        std_ = np.std(X[:,i])
        A[:,i] = (X[:,i] - mean_)/std_
    return A

Y = normalization(X)
Z = standarization(X)

```
