# Wprowadzenie

Skrypt pokazuje jak użyć pakietu SciKit do wizualizacji danych na przykładzie zestawu danych zawierającego obrazy cyfr pisanych ręcznie.

```python
% matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from sklearn import (datasets, decomposition, ensemble, manifold, random_projection)
```

```python
def plot_digits(X, y):
    X = (X - np.min(X, 0)) / (np.max(X, 0) - np.min(X, 0))

    plt.figure(figsize=(12, 9))

    for i in range(X.shape[0]):
        plt.text(X[i, 0], X[i, 1], str(y[i]), color=plt.cm.Set1(0.1*y[i]), fontdict={'weight':'bold', 'size':8})

    plt.xticks([])
    plt.yticks([])
    plt.show()
```

# Wczytanie danych

```python
digits = datasets.load_digits()
X = digits.data
y = digits.target
N, d = X.shape
```

# Wyświetlenie przykładowych cyfr

```python
M = 20
image = np.zeros((10 * M, 10 * M))
order = np.random.permutation(N)
for i in range(M):
    for j in range(M):
        image[10*i+1:10*i+9, 10*j+1:10*j+9] = X[order[i*M+j], :].reshape((8, 8))

plt.figure(figsize=(9, 9))
plt.imshow(image, cmap=plt.cm.binary)
plt.xticks([])
plt.yticks([])
plt.show()
```

# Rzut na dwie losowo wybrane osie

```python
X_projected = random_projection.SparseRandomProjection(n_components=2).fit_transform(X)
plot_digits(X_projected, digits.target)
```

# Principal Component Analysis (PCA)

```python
X_pca = decomposition.PCA(n_components=2, svd_solver='auto').fit_transform(X)
plot_digits(X_pca, digits.target)

X_pca1 = decomposition.PCA(n_components=7, svd_solver='randomized').fit_transform(X)
plot_digits(X_pca1, digits.target)

X_pca2 = decomposition.PCA(0.5).fit_transform(X)
plot_digits(X_pca2, digits.target)
```

# Kernel Principal Component Analysis (KPCA)

```python
X_kpca = decomposition.KernelPCA(n_components=2, kernel="rbf", gamma=10).fit_transform(X)
plot_digits(X_kpca, digits.target)

X_kpca1 = decomposition.KernelPCA(n_components=4).fit_transform(X)
plot_digits(X_kpca1, digits.target)

X_kpca2 = decomposition.KernelPCA(n_components=5, kernel="poly", degree=5).fit_transform(X)
plot_digits(X_kpca2, digits.target)

X_kpca3 = decomposition.KernelPCA(n_components=4, kernel="poly", degree=6).fit_transform(X)
plot_digits(X_kpca3, digits.target)

```

```python
X_kpca4 = decomposition.KernelPCA(n_components=2, kernel="cosine").fit_transform(X)
plot_digits(X_kpca4, digits.target)
```

# Isomap

```python
X_iso = manifold.Isomap(n_neighbors=30, n_components=2).fit_transform(X)
plot_digits(X_iso, digits.target)

X_iso1 = manifold.Isomap(n_neighbors=30, n_components=7).fit_transform(X)
plot_digits(X_iso1, digits.target)

X_iso2 = manifold.Isomap(n_neighbors=40, n_components=2, neighbors_algorithm = 'kd_tree').fit_transform(X)
plot_digits(X_iso2, digits.target)


for k in range(20, 101, 20):
    X_iso3 = manifold.Isomap(n_neighbors = k, n_components=2).fit_transform(X)
    plot_digits(X_iso3, digits.target)
```

# Locally Linear Embedding (LLE)

```python
X_lle = manifold.LocallyLinearEmbedding(30, n_components=2).fit_transform(X)
plot_digits(X_lle, digits.target)

X_lle1 = manifold.LocallyLinearEmbedding(30, n_components=2, reg = 0.01).fit_transform(X)
plot_digits(X_lle1, digits.target)

X_lle2 = manifold.LocallyLinearEmbedding(10, n_components=3).fit_transform(X)
plot_digits(X_lle2, digits.target)

X_lle3 = manifold.LocallyLinearEmbedding(10, n_components=3, reg = 0.01).fit_transform(X)
plot_digits(X_lle3, digits.target)

for k in range(20,101,20):
    X_lle4 = manifold.LocallyLinearEmbedding(k, n_components=4).fit_transform(X)
    plot_digits(X_lle4, digits.target)
```

# Multidimensional Scaling (MDS)

```python
X_mds = manifold.MDS(n_components=2, n_init=1, max_iter=100).fit_transform(X)
plot_digits(X_mds, digits.target)

X_mds1 = manifold.MDS(n_components=100, n_init=1, max_iter=100).fit_transform(X)
plot_digits(X_mds1, digits.target)

X_mds2 = manifold.MDS(n_components=2, n_init=1, max_iter=200).fit_transform(X)
plot_digits(X_mds2, digits.target)

X_mds3 = manifold.MDS(n_components=2, n_init=10, max_iter=100).fit_transform(X)
plot_digits(X_mds3, digits.target)

X_mds4 = manifold.MDS(n_components=2, n_init=1, metric=False).fit_transform(X)
plot_digits(X_mds4, digits.target)
```

# t-distributed Stochastic Neighbor Embedding (t-SNE)

```python
"""It is highly recommended to use another dimensionality reduction method (e.g. PCA for dense data
or TruncatedSVD for sparse data) to reduce the number of dimensions to a reasonable amount (e.g. 50) 
if the number of features is very high. This will suppress some noise and speed up the computation of 
pairwise distances between samples."""

X_tsne = manifold.TSNE(n_components=2, init='pca').fit_transform(X)
plot_digits(X_tsne, digits.target)

X_tsne3 = manifold.TSNE(n_components=2, perplexity=20, init='pca').fit_transform(X)
plot_digits(X_tsne3, digits.target)

X_tsne2 = manifold.TSNE(n_components=2, early_exaggeration=20, init='pca').fit_transform(X)
plot_digits(X_tsne2, digits.target)

X_tsne4 = manifold.TSNE(n_components=2, learning_rate=100, init='pca').fit_transform(X)
plot_digits(X_tsne4, digits.target)
```
