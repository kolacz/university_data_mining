```python
% matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import scipy.io

from sklearn import (datasets, decomposition, ensemble, manifold, random_projection)
```

```python
training = scipy.io.loadmat('ReducedImagesForTraining.mat')['images'].T
testing = scipy.io.loadmat('ReducedImagesForTesting.mat')['images'].T
```

```python
trainClass = np.floor(np.arange(training.shape[0])/5)+1
testClass = np.floor(np.arange(testing.shape[0])/2)+1
```

```python
data = np.concatenate((training, testing), axis=0)
classes = np.concatenate((trainClass,testClass))
```

```python
# nieprzetworzone dane
plt.figure(figsize=(12,12))
plt.scatter(data[: , 0], data[:, 1], c=classes)
plt.show()
```

```python
X_pca = decomposition.PCA(n_components=2).fit_transform(data)
plt.figure(figsize=(12,12))
plt.scatter(X_pca[: , 0], X_pca[:, 1], c=classes)
plt.show()
```

```python
X_kpca = decomposition.KernelPCA(n_components=2, kernel="poly", degree=5).fit_transform(data)
plt.figure(figsize=(12,12))
plt.scatter(X_kpca[: , 0], X_kpca[:, 1], c=classes)
plt.show()
```

```python
X_iso = manifold.Isomap(n_neighbors=5, n_components=2).fit_transform(data)
plt.figure(figsize=(12,12))
plt.scatter(X_iso[: , 0], X_iso[:, 1], c=classes)
plt.show()
```

```python
X_lle = manifold.LocallyLinearEmbedding(75, n_components=2).fit_transform(data)
plt.figure(figsize=(12,12))
plt.scatter(X_lle[: , 0], X_lle[:, 1], c=classes)
plt.show()
```

```python
X_mds = manifold.MDS(n_components=2, n_init=5, max_iter=100).fit_transform(data)
plt.figure(figsize=(12,12))
plt.scatter(X_mds[: , 0], X_mds[:, 1], c=classes)
plt.show()
```

```python
X_tsne = manifold.TSNE(n_components=2, init='pca').fit_transform(data)
plt.figure(figsize=(12,12))
plt.scatter(X_tsne[: , 0], X_tsne[:, 1], c=classes)
plt.show()
```
