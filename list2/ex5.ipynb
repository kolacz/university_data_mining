```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
ks = [7]
ds = range(3,10)
n = len(ks)
m = len(ds)

for d_i in range(m):
    d = ds[d_i]
    for k_i in range(n):
        count = 0
        K = ks[k_i]
        alpha = 2*math.pi/K
        # tw. cosinusów
        r = math.sqrt(d**2/(2*(1-math.cos(alpha))))
        A = np.zeros((K*1000,3)) #3 bo 0/1 kolor
        centers = np.zeros((K,2))
        for i in range(K):
            centers[i,0] = r * math.cos((math.pi/2 + 2*math.pi*i) / K)
            centers[i,1] = r * math.sin((math.pi/2 + 2*math.pi*i) / K)
        for i in range(K):
            for j in range(1000):
                x = np.random.randn() + centers[i,0]
                y = np.random.randn() + centers[i,1]
                A[j+(1000*i),0]= x
                A[j+(1000*i),1]= y
                z = (x-centers[i,0])**2 + (y-centers[i,1])**2
                for k in range(K):
                    if k != i:
                        if z > (x-centers[k,0])**2 + (y-centers[k,1])**2:
                            A[j+(1000*i),2] = 1
                            count += 1
        plt.scatter(A[:,0],A[:,1],c=A[:,2])
        plt.show()            
        print(d,count/(K*1000))
```
