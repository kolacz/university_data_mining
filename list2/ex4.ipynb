```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
d = 7
K = 7
alpha = 2*math.pi/K
# tw. cosinus√≥w
r = math.sqrt(d**2/(2*(1-math.cos(alpha))))
A = np.zeros((K*100,2))
for i in range(K):
    a = r * math.cos((math.pi/2 + 2*math.pi*i) / K)
    b = r * math.sin((math.pi/2 + 2*math.pi*i) / K)
    for j in range(100):
        A[j+(100*i),0]= np.random.randn() + a
        A[j+(100*i),1]= np.random.randn() + b
plt.plot(A[:,0], A[:,1], "r.")
plt.show()
```
