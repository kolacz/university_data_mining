```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.random.multivariate_normal([3,5],[[12,3],[3,1]],1000)
mean = np.mean(X,axis = 0)
std = np.std(X,axis = 0)

X_0 = (X - mean) / std

cov_0 = np.cov(X_0, rowvar=False)
w, v = np.linalg.eigh(cov_0) #w - wartości własne, v - wektory własne, Av[:,i] = w[i]v[:,w]

if(w[0]<w[1]):
    w[0],w[1] = w[1],w[0]
    v[:,[0,1]] = v[:,[1,0]]

X_pc = np.dot(X_0,v) * np.sqrt(1/w)
X_pc_ohne_Lambdas = np.dot(X_0,v)
```

```python
plt.axis('equal')
plt.plot(X[:,0],X[:,1],'r.')
plt.plot([mean[0] - 5 * v[0,0] * std[0], mean[0] + 5 * v[0,0] * std[0]],  [mean[1] - 5 * v[1,0] * std[1], mean[1] + 5 * v[1,0] * std[1]],'k-')
plt.plot([mean[0] - 5 * v[0,1] * std[0], mean[0] + 5 * v[0,1] * std[0]],  [mean[1] - 5 * v[1,1] * std[1], mean[1] + 5 * v[1,1] * std[1]],'k-')
plt.show()

plt.axis('equal')
plt.plot(X_0[:,0],X_0[:,1],'g.')
plt.plot([- 5 * v[0,0],  5 * v[0,0]], [- 5 * v[1,0], 5 * v[1,0]],'k-')
plt.plot([- 5 * v[0,1],  5 * v[0,1]], [- 5 * v[1,1], 5 * v[1,1]],'k-')
plt.show()

plt.axis('equal')
plt.plot(X_pc[:,0],X_pc[:,1],'b.')
plt.show()

plt.axis('equal')
plt.plot(X_pc_ohne_Lambdas[:,0], X_pc_ohne_Lambdas[:,1], 'm.')
plt.show()
```

```python
X_mean = np.mean(X,axis = 0)
X_cov  = np.cov(X.T)
X_corr = np.corrcoef(X.T)

X_0_mean = np.mean(X_0,axis = 0)
X_0_cov  = np.cov(X_0.T)
X_0_corr = np.corrcoef(X_0.T)

X_pc_mean = np.mean(X_pc,axis = 0)
X_pc_cov  = np.cov(X_pc.T)
X_pc_corr = np.corrcoef(X_pc.T)

#np.linalg.det(X_pc_corr)
#Wyznacznik macierzy korelacji jest miernikiem współliniowości (skorelowania) zmiennych objaśniających. 
#Im bliższy 1, tym niższy jest stopień wzajemnego skorelowania zmiennych objaśniających. Im bliższy 0, 
#tym siła tej korelacji większa.
#np.linalg.det(X_0_corr) - np.linalg.det(X_corr) # = ε

X_pc_ohne_Lambdas_mean = np.mean(X_pc_ohne_Lambdas,axis = 0)
X_pc_ohne_Lambdas_cov  = np.cov(X_pc_ohne_Lambdas.T)
X_pc_ohne_Lambdas_corr = np.corrcoef(X_pc_ohne_Lambdas.T)

print("E(X)")
print(X_pc_mean)
print(X_pc_ohne_Lambdas_mean,"\n")
print("Cov")
print(X_pc_cov)
print(X_pc_ohne_Lambdas_cov,"\n")
print("Corr")
print(X_pc_corr)
print(X_pc_ohne_Lambdas_corr,"\n")
```

```python
means = np.array([[-21,-2],[3,5],[27,12]])

cov = np.array([[12,3],[3,1]])

X = np.concatenate((np.random.multivariate_normal(means[0],cov,333), 
                    np.random.multivariate_normal(means[1],cov,333), 
                    np.random.multivariate_normal(means[2],cov,333)))

u = int(np.floor(np.random.rand()*3))
X = np.concatenate((X,np.random.multivariate_normal(means[u],cov,1)))
```

```python
mean = np.mean(X,axis = 0)
std = np.std(X, axis = 0)
X_0 = (X - mean) / std

cov_0 = np.cov(X_0, rowvar=False)
w, v = np.linalg.eigh(cov_0)

if(w[0]<w[1]):
    w[0],w[1] = w[1],w[0]
    v[:,[0,1]] = v[:,[1,0]]

X_pc = np.dot(X_0,v) * np.sqrt(1/w)
X_pc_ohne_Lambdas = np.dot(X_0,v)
```

```python
plt.axis('equal')
plt.plot(X[:,0],X[:,1],'r.')
plt.plot([mean[0] - 2.5 * v[0,0] * std[0], mean[0] + 2.5 * v[0,0] * std[0]],  [mean[1] - 2.5 * v[1,0] * std[1], mean[1] + 2.5 * v[1,0] * std[1]],'k-')
plt.plot([mean[0] - 2.5 * v[0,1] * std[0], mean[0] + 2.5 * v[0,1] * std[0]],  [mean[1] - 2.5 * v[1,1] * std[1], mean[1] + 2.5 * v[1,1] * std[1]],'k-')
plt.show()

plt.axis('equal')
plt.plot(X_0[:,0],X_0[:,1],'g.')
plt.plot([- 2.5 * v[0,0],  2.5 * v[0,0]], [- 2.5 * v[1,0], 2.5 * v[1,0]],'k-')
plt.plot([- 2.5 * v[0,1],  2.5 * v[0,1]], [- 2.5 * v[1,1], 2.5 * v[1,1]],'k-')
plt.show()


plt.axis('equal')
plt.plot(X_pc[:,0],X_pc[:,1],'b.')
plt.show()

plt.axis('equal')
plt.plot(X_pc_ohne_Lambdas[:,0], X_pc_ohne_Lambdas[:,1], 'm.')
plt.show()
```

```python
X_mean = np.mean(X,axis = 0)
X_cov  = np.cov(X.T)
X_corr = np.corrcoef(X.T)

X_0_mean = np.mean(X_0,axis = 0)
X_0_cov  = np.cov(X_0.T)
X_0_corr = np.corrcoef(X_0.T)

X_pc_mean = np.mean(X_pc,axis = 0)
X_pc_cov  = np.cov(X_pc.T)
X_pc_corr = np.corrcoef(X_pc.T)

X_pc_ohne_Lambdas_mean = np.mean(X_pc_ohne_Lambdas,axis = 0)
X_pc_ohne_Lambdas_cov  = np.cov(X_pc_ohne_Lambdas.T)
X_pc_ohne_Lambdas_corr = np.corrcoef(X_pc_ohne_Lambdas.T)

print("E(X)")
print(X_pc_mean)
print(X_pc_ohne_Lambdas_mean,"\n")
print("Cov")
print(X_pc_cov)
print(X_pc_ohne_Lambdas_cov,"\n")
print("Corr")
print(X_pc_corr)
print(X_pc_ohne_Lambdas_corr,"\n")
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target
    
def multiplot(X, y, labels, colors):
    N, d = X.shape

    plt.figure(figsize=(20,20))
    for i in range(d):
        for j in range(d):
            plt.subplot(d, d, i * d + j + 1)
            for label, color in zip(labels, colors):
                plt.axis('equal')
                plt.scatter(X[y == label, i], X[y == label, j], c = color)
                
    plt.show()
```

```python
X_mean = np.mean(X,axis = 0)
X_std  = np.std(X,axis = 0)
X_cov  = np.cov(X,rowvar = False)

X_0 = (X - X_mean) / X_std
cov_0 = np.cov(X_0, rowvar=False)
w, v = np.linalg.eigh(cov_0) #w - wartości własne, v - wektory własne, Av[:,i] = w[i]v[:,w]

w = w[::-1]
v[:,[0,1,2,3]] = v[:,[3,2,1,0]]

X_pc = np.dot(X_0,v) * (1/np.sqrt(w))
plt.figure()

plt.plot(X_pc[:,0],X_pc[:,1],'b.')
plt.axis('scaled')
plt.show()

#multiplot(X_pc, y, [0,1,2], "rgb")
```

```python
#restoring
X_pca = X_pc[:,:2]
w_pc = w[:2]
v_pc = v[:,:2]

d = iris.data
X_st = (d - np.mean(d, axis = 0))/np.std(d,axis = 0)
X_ohne_lambdas = X_pca * np.sqrt(w_pc)
X_res =  np.dot(X_ohne_lambdas,v_pc.T)
err = np.sum((X_res - X_st)**2)/150

print("restoring error:",err)
print("sum of the least eigenvalues:", np.sum(w[2:]))
```
